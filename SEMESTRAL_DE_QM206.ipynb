# Problema #1

Diseñar una función en Python que genere una cadena de 1000 nucleótidos y devuelva el número de ocurrencias de la secuencia “GCAT” en dicha cadena

 Realizar una rutina en Python que llame a la función anterior 10000 veces y que grafique en forma de un histograma los resultados obtenidos.

```python
import random

def generar_cadena_nucleotidos(longitud):
    nucleotidos = ['G', 'C', 'A', 'T']
    cadena = ''.join(random.choice(nucleotidos) for _ in range(longitud))
    return cadena

def contar_ocurrencias_GCAT(cadena):
    ocurrencias = cadena.count("GCAT")
    return ocurrencias

cadena_nucleotidos = generar_cadena_nucleotidos(1000)
ocurrencias_GCAT = contar_ocurrencias_GCAT(cadena_nucleotidos)

print("Cadena de nucleótidos:", cadena_nucleotidos)
print("Número de ocurrencias de la secuencia 'GCAT':", ocurrencias_GCAT)
```

```python
def ejecutar_experimento(num_experimentos):
    resultados = []
    for _ in range(num_experimentos):
        cadena_nucleotidos = generar_cadena_nucleotidos(1000)
        ocurrencias_GCAT = contar_ocurrencias_GCAT(cadena_nucleotidos)
        resultados.append(ocurrencias_GCAT)
    return resultados

def graficar_histograma(resultados):
    plt.hist(resultados, bins=max(resultados)-min(resultados)+1, align='left', edgecolor='black')
    plt.xlabel('Número de ocurrencias de la secuencia "GCAT"')
    plt.ylabel('Frecuencia')
    plt.title('Histograma de ocurrencias de la secuencia "GCAT" en 10000 experimentos')
    plt.show()

resultados_experimento = ejecutar_experimento(10000)
graficar_histograma(resultados_experimento)
```

# Problema #3

Cinética Química

Diseñe un programa en Python que calcule las concentraciones en función del tiempo de las especies químicas involucradas en el siguiente mecanismo cinético

1) A → B  → C

2) C ← A → B

** Las concentraciones iniciales y los valores de las constantes de velocidad quedan a su criterio


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
k1=3.0
dt1=0.1
t1=0
A1=1.0
P1=0.0
tiempo1=[]
AA1=[]
PP1=[]
while t1<60:
    tiempo1.append(t1)
    AA1.append(A1)
    PP1.append(P1)
    dA1=-k1*A1*dt1
    dP1=k1*A1*dt1
    A1=A1+dA1
    P1=P1+dP1
    t1=t1+dt1
    tiempo=np.array(tiempo1)
AA1=np.array(AA1)
PP1=np.array(PP1)

plt.plot(tiempo,AA1,tiempo1,PP1)
plt.xlabel("tiempo1(min)")
plt.ylabel("[Concentracion1]")
plt.title("Cinética de Reacción de Primer Orden A-->B")
plt.legend(title="Especies",loc="center right",labels=['[A]','[B]'])
plt.show()
```
